```python
with open("talon_discours.txt", "r", encoding = "utf8") as file:
    text = file.read()
```

```python
text[:50]
```

```python
from nltk.tokenize import word_tokenize
```

```python
liste_de_mots = word_tokenize(text)
```

```python
liste_de_mots
```

```python
# liste sans ponctuation 
import re
import re
import string
regex = re.compile('[%s]'%re.escape(string.punctuation)) 
liste_texte_sans_punctuation = []
for review in liste_de_mots:
    new_token = regex.sub(u'',review)
    if not new_token == u'':
        liste_texte_sans_punctuation.append(new_token)
print(liste_texte_sans_punctuation)
```

```python
# frequence 
from nltk import FreqDist
from nltk import bigrams
freqence_words = FreqDist(liste_de_mots)
words=list(freqence_words.keys())
freqwords=list(freqence_words.values())
```

```python
freqence_words.plot(50, cumulative=False) # Densité
freqence_words.plot(50, cumulative=True)
```

```python
# collocations
#print(text.collocations())
mycolloc=bigrams(['democratie', 'progres', 'corruption', 'kouande'])
print(list(mycolloc))
```

```python
# la lemmatisation
from nltk.stem import WordNetLemmatizer
lemmatiser = WordNetLemmatizer() 
print(lemmatiser.lemmatize("studying"))
```

```python
print(lemmatiser.lemmatize("studying", pos = 'v'))
```

```python
# les etiquetes grammaticals /post_tags
from nltk import pos_tag
liste_de_mots_tag = pos_tag(liste_de_mots)
```

```python
# stop words
import nltk
stpw_french = nltk.corpus.stopwords.words('french')
print(stpw_french)
```
